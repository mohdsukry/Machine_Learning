
# Import Python Libraries



```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
from sklearn.linear_model import Ridge
```

### Data Collection

```python
# New Dataset
x_new = [10.2541, 12.8890, 15.1123, 18.4567, 20.3345, 22.9012, 25.6678, 28.1234, 31.4456, 35.0000]
y_new = [25.1102, 29.4531, 33.8820, 42.1098, 44.5612, 51.0034, 55.2341, 62.8890, 68.1123, 75.4431]
```

```python

# Your coordinate data
# 1. Your Sample Data

x_data = np.array(x_new)
y_data = np.array(y_new)

```

### Define Function

```python
def create_line_plot(m,c,x_range,y_range):

    # 2. Generate points for the Equation Line
    # We use the min and max of your X data to define the start and end of the line
    x_line = np.linspace(min(x_range), max(x_range), 100)
    y_line = m * x_line + c

    y_range = y_data

    # 3. Create the Plotly Figure
    fig = go.Figure()

    # Add the Scatter Points (Original Data)
    fig.add_trace(go.Scatter(
        x=x_range, 
        y=y_range,
        mode='markers',
        name='Original Coordinates',
        marker=dict(size=10, color='blue')
    ))

    # Add the Regression Line (Your Equation)
    fig.add_trace(go.Scatter(
        x=x_line, 
        y=y_line,
        mode='lines',
        name='y - regression line',
        line=dict(color='red', width=2, dash='dash')
    ))

    # 4. Update Layout for clarity
    fig.update_layout(
        title='Interactive Scatter Plot with Regression Line',
        xaxis_title='X Coordinate',
        yaxis_title='Y Coordinate',
        template='plotly_white',
        hovermode='closest'
    )

    # Show the plot
    fig.show()
```

```python
def calculate_linear_params(x_data, y_data):
    """
    Calculates m (slope) and c (intercept) for the best-fit line y = mx + c
    """
    x = np.array(x_data)
    y = np.array(y_data)
    
    # Use NumPy's polyfit (degree 1 for linear) to find m and c
    m, c = np.polyfit(x, y, 1)
    
    return round(m, 4), round(c, 4)
```

```python
def create_line_error(my_m,my_c,x_data,y_data):

    # 2. Define y
    y_pred = my_m * x_data + my_c

    fig = go.Figure()

    # Add the Regression Line
    x_range = np.linspace(min(x_data) - 2, max(x_data) + 2, 100)
    y_range = my_m * x_range + my_c


    fig.add_trace(go.Scatter(x=x_range, y=y_range, mode='lines', name='Regression Line', line=dict(color='red', dash='dash')))

    # Add Error Lines (Residuals)
    # We loop through each point to create a separate line from actual to predicted
    for i in range(len(x_data)):
        fig.add_trace(go.Scatter(
            x=[x_data[i], x_data[i]],
            y=[y_data[i], y_pred[i]],
            mode='lines',
            line=dict(color='gray', width=1),
            showlegend=False,
            hoverinfo='none'
        ))

    # Add Actual Data Points
    fig.add_trace(go.Scatter(
        x=x_data, 
        y=y_data, 
        mode='markers', 
        name='Actual Data',
        marker=dict(size=10, color='blue')
    ))

    fig.update_layout(
        title='Scatter Plot with Error Lines (Residuals)',
        xaxis_title='X Coordinate',
        yaxis_title='Y Coordinate',
        template='plotly_white'
    )

    fig.show()
```

```python
def create_ridgeR(x_data,y_data):
    # Apply Ridge Regression (Alpha is the penalty strength)
    ridge_model = Ridge(alpha=1.0)
    ridge_model.fit(x_data.reshape(-1, 1),y_data)

    new_slope = ridge_model.coef_[0]
    new_intercept = ridge_model.intercept_

    print(f"New Equation: y = {new_slope:.4f}x + {new_intercept:.4f}")

    return new_slope, new_intercept
```

```python
def create_line_linearR_vs_ridgeR(my_m,my_c,my_m_R,my_c_R,x_data):

    # New calculated Ridge values (Example result of fit)
    # y = -0.1983x + 24.3125 (Calculated based on your 6 points)
    y_ridge_pred = my_m_R * x_data.flatten() + my_c_R

    fig = go.Figure()

    # 1. Actual Data
    fig.add_trace(go.Scatter(x=x_data.flatten(), y=y_data, mode='markers', name='Actual Data', marker=dict(color='black', size=10)))

    # 2. Your Original Line (High Error)
    y_orig = my_m * x_data.flatten() + my_c
    fig.add_trace(go.Scatter(x=x_data.flatten(), y=y_orig, mode='lines', name='Original Line', line=dict(color='red', dash='dash')))

    # 3. New Ridge Regression Line (Minimised Error)
    fig.add_trace(go.Scatter(x=x_data.flatten(), y=y_ridge_pred, mode='lines', name='Ridge Regression', line=dict(color='green', width=3)))

    fig.update_layout(title='Ridge Regression vs Original Equation', xaxis_title='X', yaxis_title='Y', template='plotly_white')
    fig.show()
```

```python
my_m, my_c = calculate_linear_params(x_data, y_data)

print(f"Calculated Slope (m): {my_m}")
print(f"Calculated Intercept (c): {my_c}")
print(f"Resulting Equation: y = {my_m}x + {my_c}")

```
